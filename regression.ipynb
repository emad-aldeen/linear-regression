```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./Fish.csv')

df.head(5)
```

```python
describe = df.describe()
info = df.info()

print(describe)
print(info)
```

```python

b = df['Weight'].values.reshape(-1,1)

y = df['Height'].values
```

```python
b_train, b_test, y_train, y_test = train_test_split( b, y, train_size=0.8, test_size=0.2 ,random_state=22)
```

```python
model = LinearRegression()
model.fit(b_train, y_train)
```

```python
y_predict = model.predict(b_test)
```

```python
plt.scatter(b_train, y_train, color='blue')
plt.plot(b_test, y_predict, color='red')
plt.xlabel('Height of the fish')
plt.ylabel('Weight of the fish')
plt.title('Training Data vs Model')
plt.show()
```

- ***so we conclose: that whenever the fish gets more weight it will incress of the height..***

```python
b = df['Width'].values.reshape(-1,1)

y = df['Height'].values

b_train, b_test, y_train, y_test = train_test_split( b, y, train_size=0.8, test_size=0.2 ,random_state=14)

model = LinearRegression()
model.fit(b_train, y_train)

y_predict = model.predict(b_test)
plt.scatter(b_train, y_train, color='blue')
plt.plot(b_test, y_predict, color='red')
plt.xlabel('Width of the fish')
plt.ylabel('Height of the fish')
plt.title('Training Data vs Model')
plt.show()
```

- ***so we conclose: that whenever the fish gets more width it will incress of the height also..***

```python
b = df['Weight'].values.reshape(-1,1)

y = df['Width'].values

b_train, b_test, y_train, y_test = train_test_split( b, y, train_size=0.8, test_size=0.2 ,random_state=35)

model = LinearRegression()
model.fit(b_train, y_train)

y_predict = model.predict(b_test)
plt.scatter(b_train, y_train, color='blue')
plt.plot(b_test, y_predict, color='red')
plt.xlabel('Weight of the fish')
plt.ylabel('Width of the fish')
plt.title('Training Data vs Model')
plt.show()
```
